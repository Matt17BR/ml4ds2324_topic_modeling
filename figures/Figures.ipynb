Word Cloud for Title

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import random

#Provided list of words with frequencies
word_freqs = {
    'network': 18,
    'neural': 17,
    'base': 15,
    'team': 15,
    'basketball': 12,
    'football': 11,
    'model': 10,
    'data': 10,
    'use': 9,
    'game': 8,
    'prediction': 8,
    'artificial': 8,
    'performance': 8,
    'injury': 7,
    'analysis': 7
}

#Custom color function for varying shades of pink
def pink_color_func(word, font_size, position, orientation, random_state=None, **kwargs):
    colors = ['#AC573D', '#F19E93', '#E8BA5A', '#FF8FAB', '#00433F']
    return random.choice(colors)

#Create a word cloud with varying shades of pink
wordcloud_T = WordCloud(width=800, height=800, 
                      background_color='white',
                      color_func=pink_color_func).generate_from_frequencies(word_freqs)

#Display the word cloud using matplotlib
plt.figure(figsize=(8, 8), facecolor=None)
plt.imshow(wordcloud_T, interpolation="bilinear")
plt.axis("off")
plt.tight_layout(pad=0)

plt.show()
wordcloud_T.to_file("wordcloud_T.png")
```
