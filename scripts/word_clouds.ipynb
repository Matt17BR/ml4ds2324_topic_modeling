### Word Cloud for Title

```python
import os
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import random

#Provided list of words with frequencies after lemmatization (from topic_modeling.ipynb)
word_freqs_T = dict((x, y) for x, y in words_lemm_T)

#Custom color function for varying shades of pink
def pink_color_func(word, font_size, position, orientation, **kwargs):
    colors = ['#AC573D', '#F19E93', '#E8BA5A', '#FF8FAB', '#00433F']
    return random.choice(colors)

def custom_wordcloud():
    return WordCloud(width=800, height=800, background_color='white',
                     color_func=pink_color_func, random_state=123)

#Create a word cloud with varying shades of pink
wordcloud_T = custom_wordcloud().generate_from_frequencies(word_freqs_T)

# Export the image
wordcloud_T.to_file(os.path.join('..','figures','wordcloud_T.png'))

#Display the word cloud using matplotlib
plt.figure(figsize=(8, 8), facecolor=None)
plt.imshow(wordcloud_T, interpolation="bilinear")
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```
